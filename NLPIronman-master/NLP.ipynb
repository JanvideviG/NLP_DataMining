# convert to panda series

```python
tweet_series = pd.Series(tweets_text)
stop_words = set(stopwords.words('english')) 
```

# tokenize and remove stop words

```python
def stopwordremoval(text):
    sent_list = text.strip().split()
    words = [word for word in sent_list if word.strip() not in stop_words and word.strip() not in punctuation and len(word.strip()) > 2]
    return (' '.join(words)).strip()
```

```python
tweet_series = tweet_series.apply(stopwordremoval)
tweet_series.head()
```
